```python
#################################
# Result Plots for FasterFVA
#################################

import numpy
import matplotlib.pyplot as plt
data = numpy.loadtxt('data.csv', delimiter = ',')
num_reactions = data[:,0]
num_metabolites = data[:,1]
LP_count_advanced = data[:,2]
LP_count_standard = data[:,3]
```

```python

################################
## Main LP Comparison Plot
################################

# generate a new axis
new_rxn_axis = numpy.arange(min(num_reactions), max(num_reactions), 1)
new_LP_axis = 2*new_rxn_axis + 1


plt.plot(new_rxn_axis, new_LP_axis, color = 'orange', linestyle = 'dashed')
plt.scatter(num_reactions, LP_count_advanced, marker = 'x', linewidths=2)
plt.xlabel('Number of reactions, $v$')
plt.ylabel('Number of LPs')
# plt.xscale('log')
# plt.yscale('log')
plt.legend([r'Standard Algorithm, $2v+1$','Proposed Algorithm'])
plt.savefig('main_plot.png', dpi = 400, facecolor = 'white')
```

```python
################################
## Relative LP Comparison Plot
################################

plt.plot([min(num_reactions), max(num_reactions)],[1,1], linestyle = 'dashed', color = 'orange')
plt.scatter(num_reactions, LP_count_standard / LP_count_advanced, marker = 'x', linewidths=2,)

# plt.scatter(num_reactions, LP_count_advanced / LP_count_advanced)
plt.xscale('log')
# plt.yscale('log')
plt.xlabel(r'Number of reactions, $v$')
plt.ylabel(r'LP Reduction Ratio, $\frac{Proposed}{Literature}$')
plt.legend(['Literature Algorithm', 'Proposed Algorithm'])
plt.savefig('main_plot.png', dpi = 400, facecolor = 'white')
```

```python
################################
## LP Asymptotic Trendline
################################

def trendline(x):
    return 3.3933*x**.8714


plt.plot(new_rxn_axis, list(map(trendline, new_rxn_axis)), linestyle = '--', color = 'orange')
# plt.plot(new_rxn_axis, 2*new_rxn_axis+1, linestyle = '--', color = 'red')
plt.scatter(num_reactions, LP_count_advanced, marker = 'x', linewidths=0.5, color = 'blue')

plt.xscale('log')
plt.yscale('log')

plt.ylabel(r'Number of LPs')
plt.xlabel(r'Number of reactions, $v$')

plt.legend(['Trendline - $3.39x^{0.8714}$'])
plt.savefig('trendline.png', dpi = 400, facecolor = 'white')
```
